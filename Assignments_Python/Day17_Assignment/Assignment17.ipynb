```python
# Question 1

class Student:
    def __init__(self, student_id, name, age, percentage):
        self.student_id = student_id
        self.name = name
        self.age = age
        self.percentage = percentage

    def display(self):
        print(f"Student ID : {self.StudentId}")
        print(f"Name  : {self.Name}")
        print(f" Age : {self.Age}")
        print(f"Percentage ID : {self.Percentage}")

    def accept(self):
        self.student_id = input("Enter Student ID: ")
        self.name = input("Enter Name: ")
        self.age = int(input("Enter Age: "))
        self.percentage = float(input("Enter Percentage: "))

    def calculate_rank(self):
        return "Rank calculation logic to be implemented."

    def __str__(self):
        return f"Student({self.student_id}, {self.name}, {self.age}, {self.percentage}%)"
```

```python
# Question 2 

class EnggStudent(Student):
    def __init__(self, student_id, name, age, percentage, branch, internal_marks):
        super().__init__(student_id, name, age, percentage)
        self.branch = branch
        self.internal_marks = internal_marks

    def display(self):
        super().display()
        print(f"Branch: {self.branch}, Internal Marks: {self.internal_marks}")

    def accept(self):
        super().accept()
        self.branch = input("Enter Branch: ")
        self.internal_marks = float(input("Enter Internal Marks: "))

    def calculate_rank(self):
        return f"Engineering Rank calculated using {self.percentage} and {self.internal_marks}."

    def __str__(self):
        return f"EnggStudent({self.student_id}, {self.name}, {self.age}, {self.percentage}%, {self.branch}, {self.internal_marks})"
```

```python
#Question 3 

class MedicalStudent(Student):
    def __init__(self, student_id, name, age, percentage, specialization, marks_of_internship):
        super().__init__(student_id, name, age, percentage)
        self.specialization = specialization
        self.marks_of_internship = marks_of_internship

    def display(self):
        super().display()
        print(f"Specialization: {self.specialization}, Internship Marks: {self.marks_of_internship}")

    def accept(self):
        super().accept()
        self.specialization = input("Enter Specialization: ")
        self.marks_of_internship = float(input("Enter Internship Marks: "))

    def calculate_rank(self):
        return f"Medical Rank calculated using {self.percentage} and {self.marks_of_internship}."

    def __str__(self):
        return f"MedicalStudent({self.student_id}, {self.name}, {self.age}, {self.percentage}%, {self.specialization}, {self.marks_of_internship})"
```

```python
#Question 4 

class College:
    def __init__(self, num_students):
        self.students = []

    def add_student(self, student):
        self.students.append(student)

    def get_student(self, student_id):
        for student in self.students:
            if student.student_id == student_id:
                return student
        return None

    def remove_student(self, student_id):
        self.students = [student for student in self.students if student.student_id != student_id]

    def __str__(self):
        return "\n".join(str(student) for student in self.students)


if __name__ == "__main__":
    college = College(5)

    engg_student = EnggStudent(1, "Alice", 21, 85.5, "Computer Science", 90)
    med_student = MedicalStudent(2, "Bob", 23, 88.0, "Surgery", 92)

    college.add_student(engg_student)
    college.add_student(med_student)

    print("College Students List:")
    print(college)

    print("\nFetching Student with ID 1:")
    print(college.get_student(1))

    college.remove_student(1)

    print("\nAfter Removing Student with ID 1:")
    print(college)
```
