```python
# Question 2 

def is_palindrome():
    num = 1
    while True:
        if str(num) == ''.join(reversed(str(num))):
            yield num
        num += 1

pallindrome = is_palindrome()
 
for i in range(20):
    print(next(pallindrome))
```

```python
def is_palindrome():
    num = 1
    while True:
        if str(num) == str(num)[::-1]:
            yield num
        num += 1

pallindrome = is_palindrome()
 
for i in range(20):
    print(next(pallindrome))
```

```python
# Question 3 Write a generator function that mimics the behavior of the range ( start, stop and step)
 

def for_range(start,stop step):
    
    start = start 
    
    while start != stop: 
        yield start
        start += step
        

for i in for_range(1,20,2):
    print(i)
        
        
    
    
```
