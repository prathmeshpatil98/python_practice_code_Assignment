```python
class Book:
    count = 0  # Static variable to track number of objects

    def __init__(self, bid, bname, price, author):
        self.bid = bid
        self.bname = bname
        self.price = price
        self.author = author
        Book.count += 1  # Increment count on each object creation

    def show_book(self):
        print(f"Book ID: {self.bid}, Book Name: {self.bname}, Price: {self.price}, Author: {self.author}")

    def __del__(self):
        print(f"Destructor called for Book ID {self.bid}")
        Book.count -= 1  # Decrement count when an object is deleted

    @staticmethod
    def get_count():
        print(f"Total Books Created: {Book.count}")


# Creating Book objects
book1 = Book(100, "Rich Dad Poor Dad", 2500, "Robert Kiyosaki")
book1.show_book()

book2 = Book(101, "Think and Grow Rich", 1800, "Napoleon Hill")
book2.show_book()

# Get count of books created
Book.get_count()

# Deleting an object
del book1

# Get updated count
Book.get_count()
```

```python
# 2

class Product:
    
    discount = 0.0
    
    def __init__(self, pid=0, pname="Unknown", price=0.0, quantity=0):
        self.pid = pid
        self.pname = pname
        self.price = price
        self.quantity = quantity
    
    def __del__(self):
        print(f"Product with ID {self.pid} is being deleted.")
    
    def show_product(self):
        print(f"Product ID: {self.pid}")
        print(f"Product Name: {self.pname}")
        print(f"Price: {self.price}")
        print(f"Quantity: {self.quantity}")
        print(f"Discount: {Product.discount}%")
    
    @staticmethod
    def set_discount(discount):
        Product.discount = discount
    
    def apply_discount(self):
        discounted_price = self.price - (self.price * Product.discount / 100)
        return discounted_price
    
    @staticmethod
    def get_discount():
        """Static method to get the current discount"""
        print(f"Current Discount: {Product.discount}%")
    


product1 = Product(201, "richdad", 80000, 5)
product2 = Product(202, "poordad", 30000, 10)
    
Product.set_discount(10) 
    
product1.show_product()
print(f"Price after discount: {product1.apply_discount()}")
    
product2.show_product()
print(f"Price after discount: {product2.apply_discount()}")
    
Product.get_discount()
    
```
